# PDR Analysis 

```python
#Importing SF parcel data that contains current use(s) and square footage. I then will merge in zoning data
```

```python
import pandas as pd
```

```python
url = 'https://data.sfgov.org/api/views/a2rp-pwkh/rows.csv?accessType=DOWNLOAD'

import requests  #  third-party library for making HTTP requests
r = requests.get(url)  # gets everything (data + metadata) that a web browser would
open('LandUse2016.csv', 'wb').write(r.content)  # saves content to disk
```

```python
# parcel level land use data. SF Open Data says updated in 2019.

sflanduse = pd.read_csv('LandUse2016.csv')
```

```python
# Looking at the data
```

```python
sflanduse.head()
```

```python
sflanduse.describe()
```

```python
# columns were cut off so adjusting to show all columns
```

```python
pd.set_option("display.max.columns", None)
sflanduse.head()
```

```python
#changing column names
```

```python
sflanduse['objectid'] = sflanduse['OBJECTID']  
sflanduse['blklot'] = sflanduse['BLKLOT']
sflanduse['bldgsf'] = sflanduse['BLDGSQFT']
sflanduse['totuses'] = sflanduse['TOTAL_USES']
sflanduse['landuse'] = sflanduse['LANDUSE']
sflanduse['pdr'] = sflanduse['PDR']
sflanduse['parcelarea'] = sflanduse['SHAPE_Area']
```

```python
sflanduse.head()
```

```python
parceldata = sflanduse[['objectid', 'the_geom', 'blklot', 'bldgsf', 'totuses','landuse','pdr','parcelarea']].copy()
```

```python
parceldata.head()
```

```python
#Total square footage of PDR uses in SF. 36 million seems very high to me.

parceldata.pdr.sum()
```

```python
#bringing in zoning data which is also parcel level

url = 'https://data.sfgov.org/api/views/acdm-wktn/rows.csv?accessType=DOWNLOAD'

import requests  #  third-party library for making HTTP requests
r = requests.get(url)  # gets everything (data + metadata) that a web browser would
open('Parcels___Active_and_Retired.csv', 'wb').write(r.content)  # saves content to disk
```

```python
sfzoning = pd.read_csv('Parcels___Active_and_Retired.csv')

low_memory = False
```

```python
sfzoning.head()
```

```python
sfzoning.zoning_code.describe()
```

```python
zoningdata = sfzoning[['blklot', 'zoning_code', 'active','shape']].copy()
```

```python
zoningdata.head()
```

```python
parcelwzoning = pd.merge(parceldata, zoningdata, on='blklot', how='left')
```

```python
parcelwzoning.head()
```

```python
parcelwzoning.describe()
```

```python
parcelwzoning['geometry'] = parcelwzoning['the_geom']
```

```python
parcelwzoning.head()
```

```python
#Getting rid of duplicate columns
```

```python
parcelwzoning2 = parcelwzoning[['objectid','blklot','bldgsf','totuses','landuse','pdr','parcelarea','zoning_code', 'active','geometry']].copy()
```

```python
parcelwzoning2.head()
```

```python
#Figuring out what Zoning_Code data type is so I can filter string values for PDR
```

```python
parcelwzoning2.info()
```

```python
#Converting object to string values for zoning_code column
```

```python
parcelwzoning2['zoning'] = parcelwzoning2['zoning_code'].astype(str)
```

```python
parcelwzoning2.head()
```

```python
# Creating a 'filteredPDR' dataframe that only includes data based on the 'zoning' column containing 'PDR' as a string value. This should filter the data to only include PDR zoned parcels.
```

```python
filteredPDR = parcelwzoning2[parcelwzoning2['zoning'].str.contains("PDR")]
```

```python
filteredPDR.head()
```

```python
filteredPDR.describe()
```

```python
#total area of PDR zoned parcels is 35.8 million

filteredPDR.parcelarea.sum()
```

```python
#Total building sf of PDR zoned parcels is 17.9 million
filteredPDR.bldgsf.sum()
```

```python
#Total sf of PDR uses in PDR zoned districts is only 7.2 million which means a lot of space is going to different things in PDR districts

filteredPDR.pdr.sum()
```

```python
#Saving results
```

```python
parcelwzoning2.to_csv(r'parcelwzoning.csv')
```

```python
filteredPDR.to_csv(r'filteredpdr.csv')
```
